```python
import re
from datetime import datetime

def parseIntoPyDate(input_text):
#   exp = "^{\"month\":(\d),\"year\":(\d+)}$"
    exp = "^{(\"month\":(\d+),)?\"year\":(\d+)}$"
#     print(input_text)
    m   = re.match(exp, str(input_text))
    if m:
        matchlen = len(m.groups())
        month_match = m[matchlen-1]
        year_match = m[matchlen]
        if month_match is None:
            month_match = 1
        # Process the date
        date_str = "{month}/{day}/{year} {hour}:{min}:{sec}".format(month=month_match,day=1,year=year_match,hour=0,min=0,sec=0)
        date_fmt = datetime.strptime(date_str, '%m/%d/%Y %H:%M:%S')
        return date_fmt
    return input_text
```

```python
df2 = df.filter(regex=('StartDate.*|EndDate.*'))
columns = df2.columns.values.tolist()
```

```python
# df2["mod."+ StartDate] = df2["StartDate"].apply(lambda x: str(parseIntoPyDate(x)) if x!= None else x)
# Format all columns
for col in columns:
    df2["mod." + col] =  df2[col].apply(lambda x: str(parseIntoPyDate(x)) if x!= None else x)
df2.head()
```

```python
# def findDuration(str_time1, str_time2):
def findDuration(x):
    str_time1 = x[0]
    str_time2 = x[1]
    # for data missing
    if str_time1 == '0' or str_time2 == '0':
        return 0
    time1 = datetime.strptime(str_time1, '%Y-%m-%d %H:%M:%S')
    time2 = datetime.strptime(str_time2, '%Y-%m-%d %H:%M:%S')    
    duration = time2 - time1
#     print("duration =>" + str(duration.days))
    return duration.days
```

```python
# Process duration and add to main dataset
# Enumerate columns
df2.head()
columns = df2.columns.values.tolist()
```

```python
for col in columns:
    if col.startswith('mod.StartDate'):
        splits = col.split('.')
        if len(splits) == 3:
            col_suffix = "." + splits[2]
        else:
            col_suffix = ""
        colname_duration = "Duration" + col_suffix
        print(colname_duration)
        # Update the original
        df[colname_duration] = df2[[col,"mod.EndDate"+ col_suffix]].apply(lambda x: findDuration(x), axis = 1)
    else:
        continue
```
